```python
def k_into_number(x):
    if 'K' in x:
        return float(x.replace('K',''))*1000
    elif 'M' in x:
        return float(x.replace('M',''))*1000000
    else:
        return float(x)
```

```python
fifa['Value']=fifa.Value.apply(k_into_number)
```

```python
fifa['Wage']=fifa.Wage.str.replace('€','')
```

```python
fifa.Wage=fifa.Wage.apply(k_into_number)
```
